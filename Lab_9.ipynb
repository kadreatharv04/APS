# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
import heapq

def dijkstra(graph, start):
    # Priority queue stores tuples: (distance, node)
    priority_queue = [(0, start)]
    distances = {node: float('inf') for node in graph}
    predecessors = {node: None for node in graph}
    distances[start] = 0

    visited = set()

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        if current_node in visited:
            continue

        visited.add(current_node)

        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight

            # If a shorter path to neighbor is found
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                predecessors[neighbor] = current_node
                heapq.heappush(priority_queue, (distance, neighbor))

    return distances, predecessors
```

```python
def reconstruct_path(predecessors, start, end):
    path = []
    current = end
    while current is not None:
        path.insert(0, current)
        if current == start:
            break
        current = predecessors[current]
    return path if path and path[0] == start else []

graph = {
    'A': {'B': 4, 'C': 2},
    'B': {'A': 4, 'C': 5, 'D': 10, 'E': 3},
    'C': {'A': 2, 'B': 5, 'D': 3, 'F': 8},
    'D': {'B': 10, 'C': 3, 'E': 2},
    'E': {'B': 3, 'D': 2, 'F': 6},
    'F': {'C': 8, 'E': 6}
}

source_nodes_to_test = ['A', 'B', 'C']

for start_node in source_nodes_to_test:
    distances, predecessors = dijkstra(graph, start_node)
    print(f"\n--- Optimized Shortest paths from source node: {start_node} ---")
    for node in graph:
        if node == start_node:
            print(f"Distance from {start_node} to {node} = 0 (Path: [{start_node}])")
            continue

        path = reconstruct_path(predecessors, start_node, node)
        if distances[node] == float('inf'):
            print(f"No path from {start_node} to {node}")
        else:
            print(f"Distance from {start_node} to {node} = {distances[node]} (Path: {' -> '.join(path)})")
```
